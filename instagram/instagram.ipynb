### ***IMPORTING UTILS***

```python
import os
from io import TextIOWrapper

def write(content:str = "test", file_name:str = "temp.txt") -> None:
	try:
		# checking and creation of text file
		if(os.path.exists(file_name) == False):
			the_file:TextIOWrapper = open(file_name, "x")
			the_file.close

		# encoding='utf-8' to catch the error with sites that have characters the charmap cant handle
		the_file:TextIOWrapper = open(file_name, "w", encoding='utf-8')
		the_file.write(str(content))
		the_file.close
	except Exception as e:
		print(f"ERROR IN {'write':20}\n{e}")
```

### ***IMPORTING THE HTMLs***
to get your followers and following HTML files
- go to your instagram page in mobile view using inspect
- go to your followers page
- scroll down to load everything
- then ctrl+s to save the page as html
- then you can rename the file whatever you wnat but this code requires it to be named followers.html

you can repeat the same thing for your following page

```python
from bs4 import BeautifulSoup as bs

# open the html file
followers_html:str = ""
following_html:str = ""

with open("followers.html", "r", encoding="utf-8") as f:
    followers_html = f.read()
with open("following.html", "r", encoding="utf-8") as f:
    following_html = f.read()

# parse it with BeautifulSoup
followers_soup:bs = bs(followers_html, "html.parser")
following_soup:bs = bs(following_html, "html.parser")
```

the span class was found manually by looking for what contained the followers name
the find tool (ctrl+f) is highly recommended

```python
from typing import List

span_class:str = "span._ap3a._aaco._aacw._aacx._aad7._aade"

spans:List[bs] = followers_soup.select(span_class)
followers_list:List[str] = [s.text for s in spans]

spans:List[bs] = following_soup.select(span_class)
following_list:List[str] = [s.text for s in spans]
```

logging and inspection

```python
write("\n".join(followers_list), "followers.txt")
print(len(followers_list))
write("\n".join(following_list), "following.txt")
print(len(following_list))
```

set operations

```python
only_following:set[str] = set(following_list.copy())
only_followers:set[str] = set(followers_list.copy())
only_both:set[str] = only_following & only_followers
```

```python
print(len(only_both))
print(only_both)
```

```python
only_following = only_following - only_both
only_followers = only_followers - only_both

print(len(only_followers))
print(only_followers)
print(len(only_following))
print(only_following)
```
